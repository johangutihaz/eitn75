## Operators
```python
Arithmetic: +       -       *       **      /       //      %
Bitwise:    <<      >>      &       |       ^       ~
Comparison: <       >       <=      >=      ==      !=
Logical:    and     not     or
```

````python
# + - * ** / // %         have been covered above. Working on numbers.
# << >> & | ^ ~           are bitwise operators, and very useful working with communication and information theory.

bin(2)  # Binary representation of the decimal 2 = 10. The '0b' infront tells us that it it binary.
````

````python
bin(2 >> 1)   # Shift the bits ONE step to the right. 0b10 -> 0b01
````

````python
bin(2 << 1)   # Shift the bits ONE step to the left. 0b10 -> 0b100
````

````python
# Bitwise AND
# 6 = 0b110         110
# 1 = 0b001        &001
#                 -----
#                  =000
bin(6 & 1)
````

````python
# Bitwise OR
# 6 = 0b110         110
# 1 = 0b001        |001
#                 -----
#                  =111
bin(6 | 1)
````

````python
# Bitwise XOR
# 6 = 0b110         110
# 3 = 0b011        ^011
#                 -----
#                  =101
bin(6 ^ 3)
````
